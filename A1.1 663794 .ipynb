Como parte del último apartado de la actividad, se realizó un análisis gráfico enfocado en una subpoblación específica, seleccionando a la población masculina. El objetivo fue explorar la relación entre el nivel de obesidad (variable categórica) y el consumo de vegetales (variable cuantitativa). Dado que el nivel de obesidad es una variable categórica, fue necesario realizar una codificación numérica para poder representarla gráficamente.

A continuación se muestra el código utilizado para generar el análisis gráfico correspondiente:

```python
# Codificar NivelDeObesidad a valores numéricos
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
data = pd.read_csv("A1.1 Obesidad.csv")

# Filtrar subpoblación masculina
df = data[data["Sexo"] == "Male"].copy()

# Codificar NivelDeObesidad a valores numéricos
categorias = df["NivelDeObesidad"].unique()
mapa = {cat: i for i, cat in enumerate(categorias)}
df["NivelNum"] = df["NivelDeObesidad"].map(mapa)

# Gráfica de dispersión
plt.figure(figsize=(10, 5))
plt.scatter(
    x=df["NivelNum"],
    y=df["ComeVegetales"],
    alpha=0.5
)

# Etiquetas y títulos
plt.xticks(
    ticks=range(len(mapa)),
    labels=mapa.keys(),
    rotation=45
)
plt.xlabel("Nivel de obesidad")
plt.ylabel("Consumo de vegetales")
plt.title("Consumo de vegetales vs Nivel de obesidad (Población masculina)")
plt.show()
```

El gráfico de dispersión obtenido muestra que el consumo de vegetales se distribuye de manera similar entre las distintas categorías de obesidad en la población masculina. No se observa una tendencia clara que indique que un mayor o menor consumo de vegetales esté directamente asociado con un nivel específico de obesidad. Los valores tienden a concentrarse en niveles intermedios de consumo independientemente de la categoría de obesidad.

En conclusión, los resultados sugieren que, aunque el consumo de vegetales es un indicador importante de hábitos alimenticios saludables, no resulta suficiente por sí solo para explicar las diferencias en los niveles de obesidad dentro de la subpoblación analizada. Esto refuerza la idea de que la obesidad es un fenómeno multifactorial, influenciado por múltiples variables como la actividad física, otros hábitos alimenticios y factores de estilo de vida.
